```
import tensorflow as tf
from tensorflow import keras
import numpy as np
from sklearn.metrics import classification_report
```

```
Y = np.zeros((1000,2)).astype(int)
filename="train_data.txt"
X = np.loadtxt(filename, delimiter=",", skiprows=1)[:,[0,1]]
#X = X.T
B = np.loadtxt(filename, delimiter=",", skiprows=1)[:,2]

for i in range(len(B)):
  if B[i]==0:
    Y[i,0] = 1
    Y[i,1] = 0
  elif B[i] == 1:
    Y[i,0] = 0
    Y[i,1] = 1

#Y = Y.T

Y_test = np.zeros((1000,2)).astype(int)
filename="test_data.txt"
X_test = np.loadtxt(filename, delimiter=",", skiprows=1)[:,[0,1]]


#X_test = X_test.T
B_test = np.loadtxt(filename, delimiter=",", skiprows=1)[:,2]

for i in range(len(B_test)):
  if B_test[i]==0:
    Y_test[i,0] = 1
    Y_test[i,1] = 0
  elif B_test[i] == 1:
    Y_test[i,0] = 0
    Y_test[i,1] = 1

#Y_test = Y_test.T
```

```
print(X.shape,Y.shape)
print('test data shapes:',X_test.shape,Y_test.shape)
```

```
X = X.reshape(1000,2,1)
Y = Y.reshape(1000,2,1)
X_test = X_test.reshape(1000,2,1)
Y_test = Y_test.reshape(1000,2,)
```

```
model = keras.Sequential([keras.layers.Flatten(input_shape=(2,1)),
                          keras.layers.Dense(3,activation="relu"),
                          keras.layers.Dense(3,activation="relu"),
                          keras.layers.Dense(2,activation="softmax")])
```

```
model.summary()
```

```
model.compile(optimizer="SGD",loss = "mean_squared_error",metrics = ["accuracy"])
```

```
X_test.T.shape
```

```
model.fit(X,Y,epochs = 500)
```

```
test_loss,test_acc = model.evaluate(X_test,Y_test)
```

```
print("Test accuracy is:",test_acc*100)
```
