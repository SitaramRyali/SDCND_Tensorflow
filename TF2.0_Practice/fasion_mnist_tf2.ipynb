```
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```
#load the fashion mnist data from the keras  dataset
data =  keras.datasets.fashion_mnist
```

###Split the data into the training and testing datasets


```
(train_images, train_labels),(test_images,test_labels) = data.load_data()
```

```
print(train_labels[6])
```

```
#create a list which defines what the label mean.
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```
print(train_images[7].shape)
plt.imshow(train_images[7],cmap = plt.cm.binary)
```

```
#normalize the images
train_images=train_images/255.0
test_images = test_images/255.0

in_shape = train_images[1].shape
```

```
in_shape
```

### Now lets create a Tensorflow model


```
model = keras.Sequential([keras.layers.Flatten(input_shape = in_shape),
                          keras.layers.Dense(128,activation="relu"),
                          keras.layers.Dense(10,activation="softmax")])
```

```
model
```

```
#create an optimizer for this model
model.compile(optimizer = "adam",loss = "sparse_categorical_crossentropy",metrics =["accuracy"])
```

```
model.fit(train_images, train_labels,epochs = 25  )
```

```
#now evaluate the model with the test data
test_loss,test_acc = model.evaluate(test_images,test_labels)
print("Tested Accuracy:",test_acc)
```

```
#now predictions
prediction = model.predict(test_images)
```
